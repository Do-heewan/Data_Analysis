```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('출생아수__출산율.xlsx',index_col='시군구별')
data
```

```python
data.info()
```

```python
# 출생아 수 추출 : birth_data 데이터프레임으로 저장
birth_data = data.bitrh_data = data.iloc[:, ::2]
birth_data
```

```python
# birth_data에서 '시군구별','전국' 행 삭제 
birth_data = birth_data.drop(['시군구별'])
birth_data = birth_data.drop(['전국'])
birth_data
```

```python
birth_data.info()
```

```python
# birth_data 모든 열을 int64로 변환
birth_data = birth_data.astype('int64')
```

```python
birth_data.info()
```

```python
# .csv 파일로 출생아 수 데이터 (birth_data) 저장하기 
birth_data.to_csv('birth_data.csv')
```

```python
# 출산율 추출 : birth_rate로저장
birth_rate = data.iloc[:, 1::2]
birth_rate
```

```python
# birth_rate '시군구별' 행 삭제
birth_rate = birth_rate.drop(['시군구별'])
birth_rate = birth_rate.drop(['전국'])
birth_rate
```

```python
birth_rate.info()
```

```python
# birth_rate 모든 열을 float로 변환
birth_rate = birth_rate.astype('float64')
birth_rate.info()
```

```python
# birth_rate에서 컬럼명에서 년도만 컬럼명으로 표시함
birth_rate.columns = [cols.split('.')[0] for cols in birth_rate.columns]
birth_rate
```

```python
# .csv 파일로 출산율 데이터(birth_rate )저장하기 
birth_rate.to_csv('birth_rate.csv')
```

```python
# 출생아 데이터(birth_data)에서 2023년만 추출하여 birth_data_2023에 저장
birth_data_2023 = birth_data['2023']
birth_data_2023
```

```python
# 출생율 데이터(birth_rate)에서 2023년만 추출하여 birth_rate_2023에 저장
birth_rate_2023 = birth_rate['2023']
birth_rate_2023
```

```python
# birth_data_2023 데이터로 시각화

import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(birth_data_2023, marker='o')
plt.xlabel('시도명')
plt.ylabel('출생아 수')
plt.title('2023 전국 출생아 수')
plt.xticks(rotation=45)  # X축 제목 45도 기울어서 표시
plt.legend()
plt.grid()
```

```python
# birth_rate_2023 데이터로 시각화

import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(birth_rate_2023, marker='o')
plt.xlabel('시도명')
plt.ylabel('출생률')
plt.title('2023 전국 출생률')
plt.xticks(rotation=45)  # X축 제목 45도 기울어서 표시
plt.legend()
plt.grid()
```

```python
# birth_rate_2023 데이터로 시각화 : 값 표시

import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(birth_rate_2023, marker='o')
plt.xlabel('시도명')
plt.ylabel('출생률')
plt.title('2023 전국 출생률')
plt.xticks(rotation=45)  # X축 제목 45도 기울어서 표시
plt.legend()
plt.grid()

for i, value in enumerate(birth_rate_2023):
    plt.text(i, value, '%.2f' % value, fontsize=10, ha='center', va='bottom' )
```

```python
# birth_rate_2023 데이터로 시각화 : bar

plt.figure(figsize=(15, 5))
plt.bar(birth_rate_2023.index, birth_rate_2023, color = 'g')
plt.xlabel('시도명')
plt.ylabel('출생률')
plt.title('2023 전국 출생률')
plt.xticks(rotation=45)  # X축 제목 45도 기울어서 표시
plt.legend()
plt.grid()
```

```python
# birth_rate_2023 데이터로 시각화 : bar, 데이터 값 표시

plt.figure(figsize=(15, 5))
bar = plt.bar(birth_rate_2023.index, birth_rate_2023, color = 'g')
plt.xlabel('시도명')
plt.ylabel('출생률')
plt.title('2023 전국 출생률')
plt.xticks(rotation=45)  # X축 제목 45도 기울어서 표시
plt.legend()
plt.grid()

for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.2f' % height, ha='center', va='bottom', size = 12)
```
