## 광역시도의 인구 비율 그래프 그리기(파이, 막대)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family ='Malgun Gothic') # matplotlib에서 한글 표시 되도록 설정
```

```python
df = pd.read_csv('광역시도.csv') 
df.rename(columns={'행정구역':'시도명'}, inplace=True)
df
```

- filter() 함수 : 지정된 인덱스 레이블에 따라 데이터 프레임 행 또는 열을 부분 집합화
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.filter.html?highlight=filter#pandas.DataFrame.filter

```python
df1 = df.filter(['시도명'])
df1
```

### 나이대를 '0-9세', '10대',.....'90대', '100세이상' 으로 변경
- 0-9세, 10-19세, ..... 90-99세 나이대의 합을 구한다

```python
df1['0-9세']=df.loc[ :  ,"0세":"9세"].sum(axis=1)
df1['10대']=df.loc[ : , "10세" : "19세"].sum(axis=1)
df1['20대']=df.loc[ : , "20세" : "29세"].sum(axis=1)
df1['30대']=df.loc[ : , "30세" : "39세"].sum(axis=1)
df1['40대']=df.loc[ : , "40세" : "49세"].sum(axis=1)
df1['50대']=df.loc[ : , "50세" : "59세"].sum(axis=1)                            
df1['60대']=df.loc[ : , "60세" : "69세"].sum(axis=1)                       
df1['70대']=df.loc[ : , "70세" : "79세"].sum(axis=1)
df1['80대']=df.loc[ : , "80세" : "89세"].sum(axis=1)
df1['90대']=df.loc[ : , "90세" : "99세"].sum(axis=1)
df1['100세이상']=df.filter(['100세 이상'])
df1
```

```python
df2 = df1.set_index('시도명')
```

```python
df2
```

#### 아래 코드를 이용하여 '광역시도명'별 나이대의 인구 비율(%)을 비교 그래프에 텍스트를 표시하시오

```python
import matplotlib.pyplot as plt

# 샘플 데이터
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 15, 30, 25]

# 가로 막대 그래프 생성
plt.barh(categories, values)

# 값들을 텍스트 라벨로 추가
for i, v in enumerate(values):
    plt.text(v, i, str(v), va='center')

# 그래프 커스터마이징
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart with Value Labels')

# 그래프 보여주기
plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터 예시
categories = ['A', 'B', 'C', 'D']
values = [10, -15, 20, -25]

# 수평 막대 그래프 그리기
plt.barh(categories, values)

# 텍스트 표시
for i, v in enumerate(values):
    plt.text(v, i, str(v), ha='left' if v >= 0 else 'right')

plt.show()
```
