#### 다양한 데이터를 읽고 저장하기
- .csv, .xlsx, .json, .html 데이터 읽기(read), 저장(쓰기)

```python
import pandas as pd
person_data = {'성명' : ['이순신', '김유진'], 
            '나이' : [22, 19], '성별' : ['남', '여'],
            '키' : [175, 163], '체중' : [75, 53]}
df = pd.DataFrame(person_data)
df
```

```python
# .csv 파일로 저장하기
df.to_csv('person.csv')
```

```python
# .csv 파일 읽기
df = pd.read_csv('person.csv')
df
```

```python
# .csv 파일 읽기, 인덱스 0 열을 무시
df1 = pd.read_csv('person.csv', index_col=0)
df1
```

```python
# .csv 파일로 저장하기 : 인덱스를 무시하고 저장
df.to_csv('person11.csv', index = False)
```

```python
# .csv 파일 읽기
df11 = pd.read_csv('person11.csv')
df11
```

#### 대한민국 출생율 & 사망율

```python
import pandas as pd
url = "https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%B8%EA%B5%AC" #대한민국 인구
wiki_tables = pd.read_html(url)   
wiki1 = wiki_tables[5]
wiki1
```

```python
# .csv 파일로 저장하기 : 인덱스 무시
wiki1.to_csv('출생_사망 인구.csv', index = False)
```

```python
# .csv 파일 읽기
wiki1_ingu = pd.read_csv('출생_사망 인구.csv')
wiki1_ingu
```
