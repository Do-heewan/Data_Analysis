```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('출생아수__출산율.xlsx',index_col='시군구별')
data
```

```python
data.info()
```

```python
# 출생아 수 추출
birth_data = data.iloc[:,0::2] 
birth_data
```

```python
# # '시군구별' 행 삭제
# birth_data = birth_data.drop('시군구별')
# birth_data
# '시군구별','전국' 행 삭제
birth_data = birth_data.drop(['시군구별','전국'])
birth_data
```

```python
birth_data.info()
```

```python
# 출산율 추출
birth_rate = data.iloc[:,1::2]
birth_rate
```

```python
# '시군구별' 행 삭제
birth_rate = birth_rate.drop(['시군구별','전국'])
birth_rate
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 4))
plt.plot(birth_rate_2023, color='b', marker='o')

# 데이터 값 표시
for i, value in enumerate(birth_rate_2023):
    plt.text(i, value, '%.2f' % value, fontsize=10, ha='center', va='bottom')  # 데이터 값을 표시

plt.xlabel('시도명')
plt.ylabel('출산율')
plt.title('2023 전국 출산율')
plt.xticks(rotation=45)  # X축 제목 45도 기울어서 표시
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,4))
bar = plt.bar(birth_rate_2023.index, birth_rate_2023, color='g')
#plt.ylim(0.5,1.5)
plt.xlabel('시도명')
plt.ylabel('출산율')
plt.title('2023년 전국 출산율')
plt.xticks(rotation=45) # X출 제목 45도 기울어서 표시
plt.grid()
# 숫자 넣는 부분
for rect in bar:    
    height = rect.get_height() # bar(막대)의 y 값을 height에 저장  
    plt.text(rect.get_x() + rect.get_width()/2.0, height,
            '%.2f' % height, ha='center', va='bottom', size = 12)   
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
for year in birth_rate.columns:    
    plt.plot(birth_rate[year], marker='o', label=year)

plt.xlabel('시도명')
plt.ylabel('출산율')
plt.title('2017년 ~ 2023년 전국 출산율')
plt.xticks(rotation=45)  # X축 제목 45도 기울어서 표시
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
# 출산율 데이터를 저장할 리스트
max_values = []
years = birth_rate.columns
for year in years:    
    chartdata = birth_rate[year]
    plt.plot(chartdata, marker='o', label=year)
    
    # 최대 출산율 및 해당 시도명 찾기
    max_value = chartdata.max()
    max_index = chartdata.idxmax()  # 최대값의 인덱스(시도명) 찾기
    max_values.append((max_index, max_value))

    # 최대값 위치에 데이터 텍스트 표시
    plt.text(chartdata.index.get_loc(max_index), max_value, f'{max_value:.2f}', 
             fontsize=10, ha='center', va='bottom', color='red')

plt.xlabel('시도명')
plt.ylabel('출산율')
plt.title('2017년 ~ 2023년 전국 출산율', fontsize=16)
plt.xticks(rotation=45)  # X축 제목 45도 기울어서 표시
plt.legend()
plt.grid()
plt.show()
```
